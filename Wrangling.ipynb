```python
# drop country column
df.drop(['country'], axis=1, inplace=True)
```

```python
# sort rows based on "id' in ascending Order
sort_df = df.sort_values(by = 'id', ascending = True)
```

```python
#rename seasons
sort_df['season'] = sort_df['season'].replace({'dry and plenty season of agric produce': 'dry & plenty', 
                                               'lean and wet season - high availability of WEP': 'lean & wet'})
```

```python
#confirm
sort_df['season'].value_counts()
```

```python
#rename columns
sort_df.rename(columns = {'proteintotal_g':'Protein(g)', 'fattotal_g':'fat(g)',
                              'chototal_g':'carbohydrate(g)', 'ca_mg':'Calcium(g)',
                         'fe_mg':'Iron(g)', 'zn_mg':'Zinc(g)', 'vitaraemcg':'VitaminA(mcg)',
                         'foodweight':'FoodWeight(g)', 'folate_mcg':'Folate(mcg)', 'energy_kcal':'Energy(Kcal)',
                         'vitamina_RE':'VitaminA(RE)', 'vitaminc_mg':'VitaminC(mcg)',
                         'speciescount':'SpeciesCount'}, inplace = True)
```

```python
#round to 3 decimal places
sort_df.round(3)
```

```python
#rename columns
sort_df.rename(columns = {'vitamimanar':'VitA_NAR', 'vitamimcnar':'VitC_NAR', 'zn_mgnar':'Zn_NAR',
                         'ca_mgnar':'Ca_NAR', 'fe_mgnar':'Fe_NAR', 'folate_mcgnar':'Folate_NAR',
                         'MAR':'Mean_NAR'}, inplace = True)
```

```python
#Sort children data
Children = sort_df[(sort_df['age'] <= 23) & (sort_df['age'] >= 12)]
```

```python
#Sort adult data
Adult = sort_df[(sort_df['age'] >= 24)]
```

```python
#Store children data to CSV
Children.to_csv('children.csv', index = False)
```

```python
#Store adult data to CSV
Adult.to_csv('adult.csv', index = False)
```
